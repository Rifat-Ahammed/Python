# Reading Json to CSV

```python
import numpy as np
import pandas as pd
```

```python
data = '{"employee": "James", "email":"james@gmail.com", "job_profile": [{"title1":"Team Lead", "title2": "Sr. Developer"}]}'
```

```python
df = pd.read_json(data)
```

```python
df
```

```python
df.to_json()
```

```python
df.to_json(orient='records') # difference between using orient parameter.
```
