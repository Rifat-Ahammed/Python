### plt.plot

```python
y = x * x
```

```python
## plt.plot

plt.plot(x, y, 'r*', linestyle='dashed', linewidth=2, markersize=12)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2D Diagram')
```

```python
## creating subplot

plt.subplot(2, 2, 1)
plt.plot(x, y, 'r--')
plt.subplot(2, 2, 2)
plt.plot(x, y, 'g*--')
plt.subplot(2, 2, 3)
plt.plot(x, y, 'bo')
```

## Sine wave form using matplotlib

```python
np.pi
```

```python
# compute the x and y coordinates for points on a sine curve

x = np.arange(0, 4 * np.pi, 0.1)
y = np.sin(x)
```

```python
plt.title('Sine Wave Form')
plt.plot(x, y, '--')
plt.xlabel('X axis')
plt.ylabel('Y Axis')
```

```python
## subplot

# compute the x and y coorinates for points on sine and cosine curves

x = np.arange(0, 5*np.pi, 0.1)

y_sin = np.sin(x)
y_cos = np.cos(x)
```

```python
# make the first plot
plt.subplot(2, 1, 1)

plt.title('Sine')
plt.plot(x, y_sin, 'r--')

# make the second plot

plt.subplot(2, 1, 2)
plt.title('Cosine')
plt.plot(x, y_cos, 'g--')
```

## Bar plot

```python
x = [2, 8, 10]
y = [11, 16, 9]

x2 = [3, 9, 11]
y2 = [6, 15, 7]

plt.bar(x, y, align='center')
plt.bar(x2, y2, color='g')
plt.title('Bar graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
```

## Box Plot uisng matplotlib

A Box Plot is also known as Whisker plot is created to display the summary of the set of data values having properties like minimum, first quartile, median, third quartile and maximum.  Here x-axis denotes the data to be plotted while the y-axis shows the frequency distribution.

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]
```

```python
plt.boxplot(data, vert=True, patch_artist=True)
```

## Pie Chart

```python
# Data to plot

labels = 'Python', 'C++', 'Java', 'Ruby'
sizes = [215, 130, 245, 210]
colors = ['gold','yellowgreen', 'lightcoral', 'lightskyblue']
explode = [0.1, 0, 0, 0] # explode first slice

#plot

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct = '%1.1f%%', shadow=True)
plt.axis('equal')
```
