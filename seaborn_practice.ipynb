```python
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = sns.load_dataset('iris')
```

```python
df.corr()
```

## Heatmap

```python
sns.heatmap(df.corr(), cmap='RdBu')
```

## JoinPlot

```python
sns.jointplot(x = 'sepal_length', y = 'sepal_width', data = df, kind='hex')
```

```python
sns.jointplot(x ='sepal_width', y='petal_width', data=df, kind='reg')
```

## Pairpolot

```python
 sns.pairplot(df, hue='sepal_length')
```

## Distplot

```python
sns.distplot(df['sepal_width'])
```

```python
sns.distplot(df['petal_length'], kde=False, bins=10)
```

## Countplot

```python
sns.countplot(x = 'sepal_length', data = df)
```

## BarPlot

```python
sns.barplot(x = 'sepal_length', y = 'petal_width', data = df)
```

## BoxPlot

```python
sns.boxplot(x = 'sepal_width', y = '')
```
